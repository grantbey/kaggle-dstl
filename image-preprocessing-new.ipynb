```python
# Apply histogram equilization

def hist_eq(inputs):
    imgs = inputs.copy()
    for img in range(imgs.shape[0]):
        for channel in range(imgs.shape[1]):
            image = imgs[img,channel,...]
            image = skimage.exposure.equalize_hist(image)
            imgs[img,channel,...] = image
    return imgs

# Align images using first three channels as anchor

def aligner(inputs):
    
    def alignment(anchor,img):

        anchor = cv2.cvtColor(np.rollaxis(anchor,0,3),cv2.COLOR_BGR2GRAY)
        img = img[:,:,np.newaxis]
        # lp1 = cv2.Laplacian(p1,cv2.CV_32F,ksize=5)
        # lp2 = cv2.Laplacian(p2,cv2.CV_32F,ksize=5)

        # Define the motion model
        warp_mode = cv2.MOTION_EUCLIDEAN
        #warp_mode = cv2.MOTION_TRANSLATION

        # Define 2x3 matrices and initialize the matrix to identity
        warp_matrix = np.eye(2, 3, dtype=np.float32) # In OpenCV an Affine transform is stored in a 2 x 3 sized matrix

        # Specify the number of iterations.
        # number_of_iterations = 5000 # Uses 1000 below

        # Specify the threshold of the increment
        # in the correlation coefficient between two iterations
        # termination_eps = 1e-10 # uses 1e-7 below

        # Define termination criteria
        criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 1000,  1e-7)

        # Run the ECC algorithm. The results are stored in warp_matrix.
        try:
            (cc, warp_matrix) = cv2.findTransformECC(anchor, img,warp_matrix, warp_mode, criteria)
            #print("_align_two_rasters: cc:{}".format(cc))
        except Exception:
            try:
                criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10000,  1e-7)
                (cc, warp_matrix) = cv2.findTransformECC(anchor, img,warp_matrix, warp_mode, criteria)
                #print("cc:{}".format(cc))
            except:
                #print('Image %i, channel %i failed to converge' %(i, channel))
                pass   

        # Use warpAffine for Translation, Euclidean and Affine
        output = cv2.warpAffine(img, warp_matrix, (anchor.shape[1], anchor.shape[0]), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
        output[output == 0] = np.average(output)

        return output
    
    aligned = inputs.copy()
    for i in range(inputs.shape[0]):
        for channel in range(3,20):
            aligned[i,channel,...] = alignment(inputs[i,:3,...],inputs[i,channel,...])
        #print('Image %i done.\n' %i)
    return aligned

# Normalize pixel values between 0,1
def min_max_scaler(inputs):
    for i in range(inputs.shape[0]):
        for j in range(inputs.shape[1]):
            inputs[i,j,:,:] = (inputs[i,j,:,:].min() - inputs[i,j,:,:])/(inputs[i,j,:,:].min() - inputs[i,j,:,:].max())
    return inputs

# Normalize pixel values between -1,1
def minus_one_scaler(inputs):
    for i in range(inputs.shape[0]):
        for j in range(inputs.shape[1]):
            inputs[i,j,:,:] = (inputs[i,j,:,:].min() - inputs[i,j,:,:])/(inputs[i,j,:,:].min() - inputs[i,j,:,:].max()) * 2 -1
    return inputs

# Normalize pixel values between 0,1
def zero_mean_scaler(inputs):
    for i in range(inputs.shape[0]):
        for j in range(inputs.shape[1]):
            inputs[i,j,...] = (inputs[i,j,...] - inputs[i,j,...].mean())/(inputs[i,j,...].std())
    return inputs

# ZCA whitening
def zca_whitening(inputs):
    sigma = np.dot(inputs, inputs.T)/inputs.shape[1] #Correlation matrix
    U,S,V = np.linalg.svd(sigma) #Singular Value Decomposition
    epsilon = 0.1 #Whitening constant, it prevents division by zero
    ZCAMatrix = np.dot(np.dot(U, np.diag(1.0/np.sqrt(np.diag(S) + epsilon))), U.T) #ZCA Whitening matrix
    return np.dot(ZCAMatrix, inputs) #Data whitening

# Here we send the x array to the ZCA whitening function and save the result in a new array
def zca(inputs):
    x_whitened = np.empty(inputs.shape)
    img_h,img_w = 136,136
    for i in range(inputs.shape[0]):
        for j in range(inputs.shape[1]):
            img_h,img_w = inputs.shape[2:4]
            x_whitened[i,j,:,:] = zca_whitening(inputs[i,j,:,:].reshape(-1,img_h*img_w)).reshape(img_h,img_w)
    
    return x_whitened
```

```python
# This cell processes each submission image
# The results are stored in a new dict, which is subsequently saved like before using np.save()

proc_sub_imgs = {}
for i,imageID in enumerate(submission_imgs.keys()):
    img = submission_imgs[imageID]
    img = img[np.newaxis,...]
    img = minus_one_scaler(img)
    img = hist_eq(img)
    img = zero_mean_scaler(img)
    img = zca(img)
    img = aligner(img.astype(np.float32))
    img = minus_one_scaler(img)
    print('{} of {}: {}'.format(i,len(submission_imgs),imageID))
    proc_sub_imgs[imageID] = img
push()
```

```python
# plot RGB
# Note: must be normed to 0,1 otherwise all black
x_min_max = min_max_scaler(x_train)
plt.imshow(np.rollaxis(x_min_max,1,4)[5,:,:,:3])
plt.show()
```

```python
# Perform histogram equalization on images scaled to (0,1)

x_minus_one_scaled_hist_eq = hist_eq(minus_one_scaler(x_train))
plt.imshow(np.rollaxis(x_minus_one_scaled_hist_eq,1,4)[5,:,:,0],cmap='Greys')
plt.show()
```

**Note:** at this point, x_min_max is regular data but scaled to (0,1), whilst x_minus_one_scaled_hist_eq is histogram equalized on data scaled to (-1,1).

```python
# Perform zero mean unit variance scaling on each channel

x_minus_one_scaled_hist_eq_zero_mean = zero_mean_scaler(x_minus_one_scaled_hist_eq)
plt.imshow(np.rollaxis(x_minus_one_scaled_hist_eq_zero_mean,1,4)[5,:,:,0],cmap='Greys')
plt.show()
```

```python
# Perform ZCA whitening on the scaled, hist_eq'd and zero-mean'd data

x_minus_one_scaled_hist_eq_zero_mean_zca = zca(x_minus_one_scaled_hist_eq_zero_mean)
plt.imshow(np.rollaxis(x_minus_one_scaled_hist_eq_zero_mean_zca,1,4)[5,:,:,0],cmap='Greys')
plt.show()
```
